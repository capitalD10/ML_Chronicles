```python
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

**Loading the Dataset**

```python
df = sns.load_dataset("mpg")
print(df.head())
```

```python
print(df.info())
```

```python
df.drop(["name"], axis = 1, inplace = True)

# removing the "name" table from the dataset (because unnecessary)
# setting axis to 1 -> to remove it column by column
# axis=0 denotes rows and axis=1 denotes columns (for searching in the table)
# inplace = True parameter makes it permanent
```

```python
print(df.head())
```

**Summary of the Dataset**

```python
print(df.describe())
```

**Data Preprocessing**

```python
print(df.isnull().sum())
# sees if there are any null values in the dataset
# sum() sums all the null values into one
# at first horsepower value was 6, after dropping it became 0
```

```python
print(df[df.isnull().any(axis = 1)])
# [] is sused for indexing
# if there is at least one empty value in the column, it will give an output
```

```python
df.dropna(inplace = True)
# the null values are droppped here
```

```python
print(df.isnull().sum().any())
# False means there are no null values and vice-versa
```
